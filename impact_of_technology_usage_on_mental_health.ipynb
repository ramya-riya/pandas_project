```python
import numpy as np
def replace_outliers(dataframe,feature):
    Q1=dataframe[feature].quantile(0.25)
    Q3=dataframe[feature].quantile(0.75)
    IQR=Q3-Q1
    a=Q1-1.5*IQR  # lower
    b=Q3+1.5*IQR

    dataframe[feature]=np.where(dataframe[feature]>b,b,dataframe[feature])
    dataframe[feature]=np.where(dataframe[feature]<a,a,dataframe[feature])
```

```python
replace_outliers(tech_usage,"Screen_Time_Hours")
```

```python
#Which age group has the highest share in the dataset?
blue_palette = ["skyblue","steelblue","deepskyblue"]
age_count=tech_usage["Age_Group"].value_counts()
plt.figure(figsize=(6,6))
age_count.plot(
    kind="pie",
    autopct='%1.1f%%',
    colors=blue_palette[:len(tech_usage["Age_Group"].unique())],
    startangle=90,
    wedgeprops={"edgecolor":"white"},
    shadow=True,
    explode=[0.03]*len(age_count)
)
plt.title("Age Group Distribution", fontsize=14, weight="bold")
plt.ylabel("")
plt.show()
```

```python
#Which age group spends the most time on screens?
age_screen=tech_usage.groupby("Age_Group")["Screen_Time_Hours"].mean().sort_values(ascending=False)

print(age_screen)
```

```python
sns.barplot(x=age_screen.index,y=age_screen.values,width=0.5)
plt.title("Screen Time vs Age Group",fontsize=14,weight="bold")
plt.xlabel("Age Group")
plt.ylabel("Average Hours on Screen Time")
plt.xticks(rotation=30)
plt.show()
```

```python
#Which age range shows the lowest sleep hours?
sleep_avg = tech_usage.groupby("Age")["Sleep_Hours"].mean().sort_index()

plt.figure(figsize=(10,5))
plt.plot(sleep_avg.index, sleep_avg.values, marker="o", color="blue")
plt.title("Average Sleep Hours by Age",fontsize=14,weight="bold")
plt.xlabel("Age")
plt.ylabel("Sleep Hours")
plt.grid(True)
plt.show()
```

```python
#Which gender group is most represented in the dataset?
gender_counts=tech_usage["Gender"].value_counts()
blue_palette=["skyblue","steelblue","deepskyblue"]
plt.figure(figsize=(6,6))
gender_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    colors=blue_palette,
    startangle=90,
    shadow=True,
    explode=[0.03]*len(gender_counts)
)

plt.title("Gender Distribution", fontsize=14, fontweight="bold")
plt.ylabel("")
plt.axis("equal")
plt.show()
```

```python
#Which gender is most active physically?
gender_activity=tech_usage.groupby("Gender")["Physical_Activity_Hours"].mean().sort_values(ascending=False)
print(gender_activity)
plt.figure(figsize=(6,5))
ax = sns.barplot(x=gender_activity.index, y=gender_activity.values, width=0.5)
for i, v in enumerate(gender_activity.values):
    ax.text(i, v + 0.1, f"{v:.1f} hrs",ha='center',fontweight="bold",color="black")

plt.title("Physical Activity Hours by Gender", fontsize=14, weight="bold")
plt.xlabel("Gender")
plt.ylabel("Avg. Physical Activity Hours")
plt.show()
```

```python
min_sleep = tech_usage["Sleep_Hours"].min()
print(min_sleep)
max_sleep = tech_usage["Sleep_Hours"].max()
print(max_sleep)
```

```python
#Which gender gets slightly more sleep?
Gender_Sleep = tech_usage.groupby("Gender")["Sleep_Hours"].mean().sort_values(ascending=False)
print(Gender_Sleep)
```

```python
num_genders = len(Gender_Sleep)
colors = sns.color_palette("viridis", num_genders)
plt.figure(figsize=(6,6))
Gender_Sleep.plot(
    kind="pie",
    autopct="%1.1f%%",
    colors=colors,
    startangle=90,
    shadow=True,
    explode=[0.03]*num_genders
)

# Title and style
plt.title("Average Sleep Hours by Gender", fontsize=14, fontweight="bold")
plt.ylabel("")
plt.axis('equal')
plt.show()
```

```python
#Is there a clear relationship between sleep and stress?
sns.scatterplot(x="Sleep_Hours",y="Stress_Level",hue="Gender",data=tech_usage,palette="Set1")
plt.title("Sleep Hours vs Stress Level",fontsize=14,weight="bold")
plt.show()
```

```python
#How do sleep hours vary with mental health status and stress level?
sleep_avg = tech_usage.groupby(["Mental_Health_Status","Stress_Level"])["Sleep_Hours"].mean()
print(sleep_avg)
```

```python
sleep_avg.plot(kind="bar",color=['#004D40', '#00796B', '#26A69A', '#80CBC4'])

plt.title("Average Sleep Hours by Mental Health and Stress Level",fontsize=14,fontweight="bold")
plt.xlabel("Mental Health Status")
plt.ylabel("Average Sleep Hours")
plt.xticks(rotation=0)
plt.legend(title="Stress Level")
plt.show()
```

```python
#How does work environment impact stress levels?
sns.countplot(x="Work_Environment_Impact",hue="Stress_Level",data=tech_usage,palette="magma")
plt.title("Work Environment Impact on Stress Levels",fontsize=14,weight="bold")
plt.show()
```

```python
#Do most people have access to support systems?
support_counts = tech_usage["Support_Systems_Access"].value_counts()
colors = ["#87CEEB", "#4682B4"]
plt.figure(figsize=(6,6))
wedges, texts, autotexts = plt.pie(
    support_counts,
    labels=support_counts.index,
    autopct='%1.1f%%',
    colors=colors,
    startangle=90,
    shadow=True,
    textprops={"fontsize": 12}
)
for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color("white")
for text in texts:
    text.set_fontsize(12)

plt.title("Support Systems Access",fontsize=14,weight="bold")
plt.show()
```

```python
#What is the overall distribution of mental health status among individuals?
mental_counts = tech_usage["Mental_Health_Status"].value_counts()
sns.barplot(x=mental_counts.index, y=mental_counts.values, width=0.5)
plt.title("Mental Health Status Distribution", fontsize=14, weight="bold")
plt.xlabel("Mental Health Status")
plt.ylabel("Number of Users")
plt.show()
```

```python
#Which daily activity takes up the largest share of average hours, and which takes the least?
plt.figure(figsize=(6,7))
colors = sns.color_palette("Set2", 3)
avg_hours = tech_usage[['Technology_Usage_Hours', 'Social_Media_Usage_Hours', 'Gaming_Hours']].mean()
labels = ['Technology Usage', 'Social Media Usage', 'Gaming']
plt.pie(avg_hours, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90, wedgeprops={'width':0.5})
plt.title("Proportion of Average Hours Spent on Activities", fontsize=14, fontweight='bold')
plt.show()
```

```python
#Is there a relationship between support system access and stress levels?
plt.figure(figsize=(8,5))
sns.countplot(
    data=tech_usage,
    x='Support_Systems_Access',
    hue='Stress_Level',
    palette='magma'
)
plt.title('Support Systems Access vs Stress Level', fontsize=14, fontweight='bold')
plt.xlabel('Support Systems Access')
plt.ylabel('Count')
plt.legend(title='Stress Level')
plt.show()
```

```python
#How does physical activity vary with mental health and stress levels?
activity_avg = tech_usage.groupby(['Mental_Health_Status', 'Stress_Level'])['Physical_Activity_Hours'].mean().sort_values(ascending=False)
print(activity_avg)
```

```python
activity_avg.plot(kind='bar',color=['#2ECC71','#F1C40F','#E67E22','#C0392B'],width=0.5)
plt.title("Average Physical Activity by Mental Health and Stress Level", fontsize=14, fontweight='bold')
plt.xlabel("Mental Health Status")
plt.ylabel("Average Physical Activity Hours")
plt.xticks(rotation=0)
plt.show()
```
